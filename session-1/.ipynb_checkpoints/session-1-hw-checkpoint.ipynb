```python
dirname = '/Users/gerbenneven/EducationalProjects/cadl/session-1/data'

filenames = [os.path.join(dirname, filename) for filename in os.listdir(dirname)]
filenames = filenames[:200:2]
```

```python
images = [plt.imread(filename)[..., :3] for filename in filenames]

images = [utils.imcrop_tosquare(img_i) for img_i in images]

images = [resize(img_i, (100, 100)) for img_i in images]

images = np.array(images).astype(np.float32)
```

# Part Two - Compute the Mean


```python
session = tf.Session()

mean_img_op = tf.reduce_mean(images, 0)
```

```python
mean_img = session.run(mean_img_op)
```

```python
mean_img.shape
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(mean_img)
plt.imsave(arr=mean_img, fname='mean.png')
```

# Part Three - Compute the Standard Deviation

```python
mean_img_4d = tf.reduce_mean(images, 0, True)
```

```python
mean_img_4d.shape
```

```python
subtraction = images - mean_img_4d

std_img_op = tf.sqrt(tf.reduce_sum(subtraction * subtraction, 0))

std_img = session.run(std_img_op)
```

```python
std_img.shape
```

```python
plt.figure(figsize=(10, 10))
std_img_show = std_img / np.max(std_img)
plt.imshow(std_img_show)
plt.imsave(arr=std_img_show, fname='std.png')
```

# Part Four - Normalize the Dataset

```python
norm_imgs_op = tf.div(tf.sub(images, mean_img), std_img)
norm_imgs = session.run(norm_imgs_op)
```

```python
norm_imgs.shape
```

```python
print(np.min(norm_imgs), np.max(norm_imgs))
print(images.dtype)
```

```python
norm_imgs.shape
```

```python
plt.figure(figsize=(10, 10))
```

```python
plt.imshow(utils.montage(norm_imgs, 'normalized.png'))
```

# Part Five - Convolve the Dataset

```python
ksize = 16
kernel = np.concatenate([utils.gabor(ksize)[:, : , np.newaxis] for i in range(3)], axis=2)
```

```python
kernel.shape
```

```python
kernel_4d = kernel.reshape([kernel.shape[0], kernel.shape[1], kernel.shape[2], 1])
assert(kernel_4d.shape == (ksize, ksize, 3, 1))
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(kernel_4d[:, :, 0, 0], cmap='gray')
plt.imsave(arr=kernel_4d[:, :, 0, 0], fname='kernel.png', cmap='gray')
```

```python
convolved = utils.convolve(norm_imgs, kernel_4d)
```

```python
convolved_show = (convolved - np.min(convolved)) / (np.max(convolved) - np.min(convolved))
print(convolved_show.shape)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(utils.montage(convolved_show[..., 0], 'convolved.png'), cmap='gray')
```

# Part six


```python
convolved.shape
```

```python
flattened = tf.reshape(convolved, [convolved.shape[0], (convolved.shape[1] * convolved.shape[2] * 1)])
```

```python
flattened.get_shape()
```

```python
assert(flattened.get_shape().as_list() == [100, 10000])
```

```python
values = tf.reduce_sum(flattened, reduction_indices=1)
```

```python
values
```

```python
idxs_op = tf.nn.top_k(values, k=100)[1]
```

```python
idxs = session.run(idxs_op)
```

```python
sorted_img = np.array([images[idx_i] for idx_i in idxs])
```

```python
assert(sorted_img.shape == (100, 100, 100, 3))
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(utils.montage(sorted_img, 'sorted.png'))
```
